```python
import pandas as pd

wine = pd.read_csv('https://bit.ly/wine_csv_data')
```

```python
wine.head()
```

```python
data = wine[['alcohol', 'sugar', 'pH']]
target = wine[['class']]
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data,target, test_size=0.2, random_state=42)
```

```python
sub_input, val_input, sub_target, val_target = train_test_split(train_input, train_target, random_state=42)
```

```python
print(data.shape, train_input.shape, test_input.shape)
print(sub_input.shape, val_input.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(sub_input, sub_target)
print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(dt, train_input, train_target)

import numpy as np
np.mean(scores['test_score'])
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
splitter = StratifiedKFold(n_splits=10, shuffle=True)
scores = cross_validate(dt, train_input, train_target, cv=splitter)
scores
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004, 0.0005]
}
```

```python
# 자동적 하이퍼파라미터를 위해 사용
dt = DecisionTreeClassifier(random_state=42)
gs = GridSearchCV(dt, params)
```

```python
gs.fit(train_input, train_target)
```

```python
gs.best_estimator_
```

```python
gs.best_params_
```

```python
gs.cv_results_
```

```python
params = {
    'min_impurity_decrease': np.arange(0.0001, 0.001, 0.0001),
    'max_depth': range(5, 20, 1),
    'min_samples_split': range(2, 100, 10)
}
```

```python
dt = DecisionTreeClassifier(random_state=42)

gs = GridSearchCV(dt, params, n_jobs=-1)
```

```python
gs.fit(train_input, train_target)
```

```python
gs.best_params_
```

### random Search

```python
from scipy.stats import uniform, randint
```

```python
rgen = randint(0, 10)
np.unique(rgen.rvs(1000), return_counts=True)
# unique = 실수생성 / randint 균등하게 숫자 생성
```

```python
params = {
    'min_impurity_decrease': uniform(0.0001, 0.001),
    'max_depth': randint(20, 50),
    'min_samples_split': randint(2, 25),
    'min_samples_leaf': randint(1, 25),
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
gs = RandomizedSearchCV(dt, params, n_iter=100, n_jobs=-1)
gs.fit(train_input, train_target)
```
